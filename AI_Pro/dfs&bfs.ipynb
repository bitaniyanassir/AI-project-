```python
class StateSpaceSearch:
    def __init__(self, graph):
        self.graph = graph

    def bfs(self, start, goals):
        visited = set()
        queue = deque([(start, [])])

        while queue:
            current_node, current_path = queue.popleft()
            if current_node in goals:
                return current_path + [current_node]
            if current_node not in visited:
                visited.add(current_node)
                for neighbor in self.graph[current_node]:
                    queue.append((neighbor, current_path + [current_node]))
        
        return None  # Path not found



search = StateSpaceSearch(graph)
```

```python
# Goals
goals = ["arba minch", "jimma", "bale", "sof oumer", "babile", "lalibela", "gonder", "axum"]

# Find paths from Addis Ababa to each goal using Breadth-First Search
print("BFS Paths:")
for goal in goals:
    bfs_path = search.bfs("addis ababa", [goal])
    if bfs_path:
       
        print(" -> ".join(bfs_path))
    else:
        print(f"No path found from addis ababa to {goal}")
```
